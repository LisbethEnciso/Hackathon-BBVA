```python
lista_ip = ["200.60.94.155:5678", "200.107.163.242:5678", "200.10.73.210:5678", "181.66.37.98:4153"]
```

```python
# Set headers
headers = requests.utils.default_headers()
headers.update({ 'User-Agent': random.choice(lista_user_agent), 'Proxies': random.choice(lista_ip)})
```

```python
lista_x = []
x = -13.159403
while x <= -12.943676:
  lista_x.append(x)
  x = x + 0.08
```

```python
lista_y = []
y = -76.446789
while y <= -76.279863:
  lista_y.append(y)
  y = y + 0.08
```

```python
len(lista_y)
```

```python
lista_lugares = [
    #'mercados',
    #'colegios',
    #'parques',
    #'bancos',
    #'comisarias',
    #'restaurantes',
    #'centros+comerciales',
    #'terminal', 
    #'cementerio',
    #'institutos', 
    #'grifos', 
    #'fabricas',
    'municipalidades',
    'hospitales'
]
```

```python
dic_f = {}
for lug in lista_lugares:
  lista = []
  for x in lista_x:
    for y in lista_y:
      url = 'https://www.google.com.pe/maps/search/'+lug+'/@'+str(x)+','+str(y)+',14.5z/data=!3m1!4b1?hl=es-419'
      lista.append(url)
      dic_f[lug] = lista
```

```python
dic_f
```

```python
def getCoordinadasMaps(url_maps, lug):
    req = requests.get(url_maps, headers)
    print(req)
    soup = BeautifulSoup(req.content, 'html.parser')
    
    texto_ = soup.text.split("[[3,-")
    
    lista_coordenadas = []

    for i, subTexto in enumerate(texto_):
        if i == 0:
            continue

        lista_coordenadas.append(('-'+subTexto[: subTexto.find(']')]).split(","))
        
    lista_coordenadas_unicas = []

    for coordenada in lista_coordenadas:
        if coordenada not in lista_coordenadas_unicas:
            lista_coordenadas_unicas.append(coordenada + [lug])

    return lista_coordenadas_unicas
```

```python
df = []
lugar = 'hospitales'
for url in dic_f[lugar]:
  df = df + getCoordinadasMaps(url, lugar)
```

```python
df = pd.DataFrame(df, columns = ['longitud', 'latitud', 'tipo_lugar'])
df = df.drop_duplicates(subset=['longitud', 'latitud', 'tipo_lugar'])
```
